```python
barometer_coordinates = client.get_sensor_coordinates(reference="proto-kraken")
```

```python
barometer_coordinates = pd.DataFrame(
    {
        "x": barometer_coordinates["geometry"]["xy_coordinates"]["x_coordinates"],
        "y": barometer_coordinates["geometry"]["xy_coordinates"]["y_coordinates"],
    }
)


barometer_coordinates["sensor"] = barometer_coordinates.index
```

```python
x_pressure_side = barometer_coordinates[barometer_coordinates["y"]<0]["x"].sort_values()
x_suction_side = barometer_coordinates[barometer_coordinates["y"]>0]["x"].sort_values()

x_suction_side
```

```python
suction_side_barometers = raw_sensor_data.dataframe.columns[x_suction_side.index]
pressure_side_barometers = raw_sensor_data.dataframe.columns[x_pressure_side.index][:-1]
```

```python
suction_side_barometers
```

```python
pressure_side_barometers
```

```python
# We should let the user enter u and p_inf in the dashboard with defaults set as:
#  u: 10 
#  p_inf: 1e5
# Then, q = 0.5 * rho * u^2 where rho = 1.225 (density of air)
```

```python
# q is dynamic pressure [Pa]
q = 365 

# p_inf is freestream pressure [Pa]
p_inf = 95.34e3
```

```python
aerodynamic_pressures = raw_sensor_data.dataframe - p_inf
aerodynamic_pressures
```

```python
cp = aerodynamic_pressures / q
cp
```

```python
# mean_p = aero.agg("mean") # These should be bined based on azimuth location and then avergad, but we can assume constant AoA and minimal wind shear for now...

suction_cp = cp[suction_side_barometers]
pressure_cp = cp[pressure_side_barometers]
```

```python
# We need to remove outliers: remove all points outside of -10 < cp < 3
```

```python
layout_dict = {
    "title_text": f"Cp Curve at {measurement_datetime}",
    "xaxis": {"title":"Chordwise sensor postition [m/1m]"},
    "yaxis": {
        "title": "Cp",
        "range": [2, -2],
        "autorange": True
    }
}


fig = px.scatter(height=800)
fig.add_scatter(x=x_pressure_side, y=pressure_cp.iloc[0], marker={"color":"red"}, mode="markers", name="Pressure Side Aerosense")
fig.add_scatter(x=x_suction_side, y=suction_cp.iloc[0], marker={"color":"blue"}, mode="markers", name="Suction Side Aerosense")
fig.update_layout(layout_dict)
fig.show()
```
